```python
dat = data3.loc[data3['Country']=='India']
dat
```

```python
training_data = data3[0:18000]
test_data = data3[18000:22000]
```

```python
training_data = dat[0:600]
test_data = dat[600:]
```

```python
independent_features = ['Country', 'WHO_region']
```

```python
!pip install pmdarima
```

```python
from pmdarima import auto_arima
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
train = training_data[training_data.columns[1:]].apply(le.fit_transform)

test = test_data[test_data.columns[1:]].apply(le.fit_transform)

train
```

```python
test
```

```python
model = auto_arima(y = train['New_deaths'],exogenous=train[independent_features],trace=True)
```

```python
model.fit(train['New_deaths'],train[independent_features])
```

```python
predict = model.predict(n_periods=len(test), exogenous=test[independent_features])
```

```python
predict
```

```python
test['Predict_ARIMA'] = predict
```

```python
test[['New_deaths','Predict_ARIMA']].plot(figsize=(15,9))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score
```

```python
np.sqrt(mean_squared_error(test['New_deaths'],test['Predict_ARIMA']))
```

```python
mean_absolute_error(test['New_deaths'],test['Predict_ARIMA'])
```
